```python
import cv2
import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import misc
from scipy import ndimage, misc
from tqdm import tqdm

image1 = cv2.imread("lena.tif")
def convert_to_gray(input1):
    r = input1[:,:,0]
    g = input1[:,:,1]
    b = input1[:,:,2]
    gray = (0.07*r)+(0.71*g)+(b*0.21)
    return gray

img = convert_to_gray(image1)
x,y = img.shape
print(x,y)
plt.imshow(img, cmap = 'gray')
```

```python
def edge_detection(image):

    x,y = image.shape
    copy_image = image.copy()
    print(x)
    print(y)
    
    for i in range(2,x-2):
        
        for j in range(2,y-2):
            
            pix_sum = 0
            p = 0
            avg = 0
            count = 0
            for k in range(-2,3):
                
                for l in range(-2,3):
                    
                    if not (k == 0 and l == 0):
                        
                        pix_sum = pix_sum + copy_image[i+k][j+l]
                        
                        if abs(copy_image[i][j]-copy_image[i+k][j+l]) <= 15:
                            
                            p = p + 1
                            
                    count = count + 1
                    
            avg = pix_sum/24
            #print(avg)
            #print(count)
            #print(copy_image[i][j])
            if abs(copy_image[i][j] -avg) <= 120 and p >= 9:
                image[i][j] = 255
            else:
                image[i][j] = 0
    return image
```

```python
out = edge_detection(img)
plt.imshow(out,cmap = 'gray')
```
