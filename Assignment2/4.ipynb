```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import time
img = cv2.imread("lena.jpg")
(x,y,z) = img.shape
```

```python
plt.imshow(img)
plt.title("Original Image")
print(img.shape)
```

```python
#part a
def RGB2YCrCb(img):
    return np.array([[[0.098*img[i,j,2]+0.504*img[i,j,1]+0.257*img[i,j,0]+16,-0.071*img[i,j,2]-0.368*img[i,j,1]+0.439*img[i,j,0]+128,0.439*img[i,j,2]-0.291*img[i,j,1]-0.148*img[i,j,0]+128]for j in range(y)] for i in range(x)]).astype(np.uint8)
```

```python
img_out = RGB2YCrCb(img)
img_cv=cv2.cvtColor(img,cv2.COLOR_RGB2YCrCb)
plt.figure(figsize=(5*3, 5*3), constrained_layout=False)
plt.subplot(1,2,1)
plt.imshow(img_cv)
plt.title('OpenCV')
plt.subplot(1,2,2)
plt.imshow(img_out)
plt.title('RGB to YCrCb converted image')
plt.show()
```

```python
#part b
def YCrCb2RGB(img):
    return np.array([[[1.164*(img[i,j,0]-16)+1.596*(img[i,j,1]-128),1.164*(img[i,j,0]-16)-0.813*(img[i,j,1]-128)-0.392*(img[i,j,2]-128),1.164*(img[i,j,0]-16)+2.017*(img[i,j,2]-128)]for j in range(y)] for i in range(x)]).astype(np.uint8)
```

```python
img_out2 = YCrCb2RGB(img_out)
img_cv2=cv2.cvtColor(img_cv,cv2.COLOR_YCrCb2RGB)
plt.figure(figsize=(5*3, 5*3), constrained_layout=False)
plt.subplot(1,2,1)
plt.imshow(img_cv2)
plt.title('OpenCV')
plt.subplot(1,2,2)
plt.imshow(img_out2)
plt.title('RGB to YCrCb converted image')
plt.show()
```
