```python
def own_function(img1, img2):
    img1 = img1.astype(np.int)
    img2 = img2.astype(np.int)
    return (img1 - img2)

def for_loop(img1, img2):
    img1 = img1.astype(np.int)
    img2 = img2.astype(np.int)
    canvas = np.zeros(img1.shape)
    for k in range(img1.shape[2]):
        for i in range(img1.shape[0]):
            for j in range(img1.shape[1]):
                if img1[i][j][k] - img2[i][j][k] <= 0: 
                    canvas[i][j][k] = 0
                else:
                    canvas[i][j][k] = img1[i][j][k] - img2[i][j][k]
    canvas = canvas.astype(np.uint8)
    return canvas
```

```python
plt.imshow(own_function(img1, img2))
```
