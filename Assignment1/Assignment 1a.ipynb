```python
#Import libraries
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage, misc
from matplotlib import pyplot as plt
import cv2
from time import time
from tqdm import tqdm
```

```python
#Read images
img1= cv2.imread('/media/harish/New Volume/IVP FIles/Assignment1/Rose.jpg')
img2= cv2.imread('/media/harish/New Volume/IVP FIles/Assignment1/lenna.tif')
```

```python
#Function to convert images to grayscale
def convert_to_gray(input1):
    r = input1[:,:,0]
    g = input1[:,:,1]
    b = input1[:,:,2]
    gray = (0.07*r)+(0.71*g)+(b*0.21)
    return gray

def for_to_gray(input2):
    prop = input2.shape
    for i in tqdm(range(prop[0]), desc = 'Converting Image', ncols = 100):
        for j in range( prop[1]):
            input2[i][j] = 0.07*input2[i][j][2] + 0.71*input2[i][j][1] + 0.21*input2[i][j][0]
    return input2

def opencv_conversion(input3):
    return cv2.cvtColor(input3, cv2.COLOR_BGR2GRAY)
```

```python
Rose_process = []
Lena_process = []
process_order = ['Own', 'For Loop', 'Built In']
```

```python
#Function calls for rose.jpg
images = []
trial = img1

# cv2.imshow('Trial', trial)
tick1 = time()
converted1 = convert_to_gray(trial)/255.
tock1 = time()
# plt.imshow(converted1, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Rose_process.append(tock1 - tick1)
images.append(converted1)

# cv2.imshow('Trial', trial)
tick1 = time()
converted2 = for_to_gray(trial)/255.
tock1 = time()
# plt.imshow(converted2, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Rose_process.append(tock1 - tick1)
images.append(converted2)

# cv2.imshow('Trial', trial)
tick1 = time()
converted3 = opencv_conversion(trial)/255.
tock1 = time()
# plt.imshow(converted3, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted3)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Rose_process.append(tock1 - tick1)
images.append(converted3)

fig, ax = plt.subplots(1, 3, figsize = (10, 10), constrained_layout = True)
for i in range(3):
    ax[i].imshow(images[i], cmap = 'gray')
#     ax[i].title(process_order[i])
#     ax[i].axes('off')
# plt.suptitle('Rose to Grayscale Conversion')
plt.show()
```

```python
#Function calls for lenna.jpg
images = []
trial = img2

# cv2.imshow('Trial', trial)
tick1 = time()
converted1 = convert_to_gray(trial)/255.
tock1 = time()
# plt.imshow(converted1, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Lena_process.append(tock1 - tick1)
images.append(converted1)

# cv2.imshow('Trial', trial)
tick1 = time()
converted2 = for_to_gray(trial)/255.
tock1 = time()
# plt.imshow(converted2, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Lena_process.append(tock1 - tick1)
images.append(converted2)

# cv2.imshow('Trial', trial)
tick1 = time()
converted3 = opencv_conversion(trial)/255.
tock1 = time()
# plt.imshow(converted3, cmap = 'gray')
# plt.show()
# cv2.imshow('converted', converted3)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
print(tock1 - tick1)
Lena_process.append(tock1 - tick1)
images.append(converted3)

fig, ax = plt.subplots(1, 3, figsize = (10, 10), constrained_layout = True)
for i in range(3):
    ax[i].imshow(images[i], cmap = 'gray')
#     ax[i].title(process_order[i])
#     ax[i].axes('off')
# plt.suptitle('Rose to Grayscale Conversion')
plt.show()
```
